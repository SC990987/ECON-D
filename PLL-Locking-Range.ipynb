```python
import numpy as np
import glob
import json
import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
import mplhep
import matplotlib.colors as mcolors

mplhep.style.use(mplhep.style.CMS)
```

```python
def jsonload(fname):
    with open(fname) as jsonfile:
        try:
            return json.load(jsonfile)
        except Exception:
            print(fname)
```

```python
fdir = "/Users/alexcampbell/Documents/Research/Fermilab/PLLLockingrange/pll_plot"
fnames = list(np.sort(glob.glob(f"{fdir}/*.json")))
all_files = fnames.copy()
```

```python
data = [jsonload(fname) for fname in fnames]

for i in range(len(data)):
    if data[i] == None:
        data.remove(data[i])
```

```python
len(data)
```

```python
locking_range_42 = []
locking_range_44 = []
for i in range(len(data)):
    for j in range(len(data[i]['tests'])):
        if "metadata" in data[i]['tests'][j]:
            locking_range_42.append(data[i]['tests'][j]['metadata']['locking_range'][0])
            locking_range_44.append(data[i]['tests'][j]['metadata']['locking_range'][1])
locking_range_42 = np.array(locking_range_42)
locking_raange_44 = np.array(locking_range_44)
```
