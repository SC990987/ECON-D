```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
import os
from ast import literal_eval
import matplotlib.colors as mcolors
import numpy
import sys
import matplotlib.patches as mpatches
numpy.set_printoptions(threshold=sys.maxsize)
import mplhep
mplhep.style.use(mplhep.style.CMS)
```

```python
def PLL2DPlot(chipNum):
    mainlist=[]
    with open(f"/Users/alexcampbell/Documents/Research/Fermilab/pll2D/plldata/chip{chipNum}.csv") as f:
        mainlist = [list(literal_eval(line)) for line in f]
    pllSettings = np.array(mainlist)
    data = pllSettings[0]
    b,a=np.meshgrid(np.arange(35,44,(1/8)),np.arange(56))
    plt.hist2d(b.flatten(),a.flatten(),weights=data.T.flatten(),bins=(np.arange(34,45,(1/8)),np.arange(57)),cmap='Blues')
    plt.xlabel('Reference Clock Frequency Setting (MHz)', size=28)
    plt.ylabel('CapBank Select Setting', size=28)
    handles, labels = plt.gca().get_legend_handles_labels()
    patch = mpatches.Patch(color='#08306b', label='PLL Lock with VCO Override')
    handles.append(patch)
    #plt.legend(handles=handles, loc='lower left')
    plt.title(f"ECON-T-P1 Chip {chipNum}")
    #plt.rcParams.update({'font.size': 15})
    plt.savefig(f'/Users/alexcampbell/Documents/Research/Fermilab/pll2D/plots/chip{chipNum}.png',dpi=300, facecolor = "w")
```

```python
PLL2DPlot(102)
```
